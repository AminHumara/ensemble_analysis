Чтобы составить спецификацию модели, в первую очередь необходимо определить признаки, влияющие на ОПЖ.  

В статьях А.Ж. Панзабековой и И.Е. Кроме того отмечаются такие факторы как численность медицинских работников, число умерших на 1000 населения и уровень бедности.  

Таким образом, выделим 6 факторов, которые влияют на ОПЖ в СКФО.

В качестве эндогенной (зависимой) переменной выступает:  
$Y_{it}$ – *ожидаемая продолжительность жизни при рождении, число лет*. Значение средней продолжительности предстоящей жизни в количестве лет, прогнозируемое в предположении, что уровни смертности населения во всех возрастах в будущем останутся такими же, как в рассматриваемом году.

В качестве экзогенных (независимых) переменных выступают:  

$X_{1_{it}}$ –  среднедушевые денежные доходы семьи (в месяц), руб. Среднедушевой доход семьи в расчетном периоде исчисляется путем деления среднемесячного совокупного дохода семьи на количество членов семьи.  
$X_{2_{it}}$ –   Доля численности населения с денежными  доходами ниже границы бедности (величины прожиточного минимума к общей численности населения субъекта, процентов  
$X_{3_{it}}$ –  Валовой региональный продукт на душу населения, руб.  
$X_{4_{it}}$ –  Число умерших на 1000 человек населения  
$X_{5_{it}}$ –  Уровень безработицы, процентов. Отношение численности безработных определенной возрастной группы к численности экономически активного населения соответствующей возрастной группы, в процентах.  
$X_{6_{it}}$ –  Численность среднего медицинского  персонала, человек (на 10 000 человек населения)  

Теперь перейдем к спецификации модели.

Для POOL спецификацию можем записать в виде:  
$$Y_{it}=\mu+a_1 \cdot X_{1_{it}}+a_2 \cdot X_{2_{it}}+a_3 \cdot X_{3_{it}}+a_4 \cdot X_{4_{it}}+a_5 \cdot X_{5_{it}}+a_6 \cdot X_{6_{it}}+\varepsilon_{it},$$
где $\mu = const, \ $ $\varepsilon_{it} - $ оцененные случайные эффекты    

Для FE спецификацию можем записать в виде:   
$$Y_{it}=\mu_i+a_1 \cdot X_{1_{it}}+a_2 \cdot X_{2_{it}}+a_3 \cdot X_{3_{it}}+a_4 \cdot X_{4_{it}}+a_5 \cdot X_{5_{it}}+a_6 \cdot X_{6_{it}}+\varepsilon_{it},$$
где	$\mu_i - $ константа для каждого субъекта, учитывающая его специфику, $\varepsilon_{it}$ – оцененные случайные эффекты.

Для модели RE спецификацию можем записать в виде:   
$$Y_{it}=\mu + a_1 \cdot X_{1_{it}}+a_2 \cdot X_{2_{it}}+a_3 \cdot X_{3_{it}}+a_4 \cdot X_{4_{it}}+a_5 \cdot X_{5_{it}}+a_6 \cdot X_{6_{it}}+ u_{i} + \varepsilon_{it},$$
где $\mu = const, \ $ $u_{i} - $ индивидуальная случайная компонента, $\varepsilon_{it}-$ остаточный член.

# Сравнение моделей

## Тест Фишера (Pool vs. FE)

image.png

image.png

```python
def pool_vs_fe(pool, fe, alpha=0.05):
    n = pool.nobs
    n_resid = fe.df_model

    RSS_fe = fe.resid_ss
    RSS_pool = (pool.resids ** 2).sum()

    F_comp = ((RSS_pool - RSS_fe) / (n - 1)) / (RSS_fe / n_resid)

    dfn = fe.df_model
    dfd = fe.df_resid

    F_crit = sts.f.ppf(1 - alpha, dfn, dfd)

    if F_comp < F_crit:
        print(f'F-значение : {F_comp:.4f} < {F_crit:.4f} ֜=> выбираем модель пула')
    else:
        print(f'F-значение : {F_comp:.4f} > {F_crit:.4f} ֜=> выбираем модель с фикс. эффектами')

pool_vs_fe(pooled_res, fe_res)
```

## Тест множителей Лагранжа (Pool vs. RE)

image.png

```python
def LM_test(df, pool, alpha=0.05):
  df_pool_resids = pd.concat([df, pool.resids], axis=1)
  df_pool_resids_mean = df_pool_resids.groupby(level=0).mean()

  ssr_grouped_resids_mean = (df_pool_resids_mean['residual']**2).sum()
  ssr_pool_resids = (pool.resids**2).sum()

  n = df_pool_resids_mean.index.size
  T = df.shape[0]/n

  bp_statistic = (n * T)/(2 * (T-1)) * ((T**2 *ssr_grouped_resids_mean)/ssr_pool_resids - 1)**2
  chi2_critical = sts.chi2.ppf((1.0-alpha), 1)

  print(f'Значение статистики : {bp_statistic:.4f}', end=' ')
  if bp_statistic < chi2_critical:
    print(f'< {chi2_critical:.4f} => выбираем модель пула')
  else:
    print(f'> {chi2_critical:.4f} => выбираем модель со случайными эффектами')

LM_test(X_pool, pooled_res, 0.05)
```

## Тест Хаусмана (FE vs. RE)

image.png

image.png

```python
def Hausman_test(fe, re, alpha=0.05):
    common_params = fe.params.index.intersection(re.params.index)
    fe_params = fe.params[common_params]
    re_params = re.params[common_params]

    fe_cov = fe.cov.loc[common_params, common_params]
    re_cov = re.cov.loc[common_params, common_params]

    delta_params = fe_params - re_params
    delta_cov = fe_cov - re_cov

    hausman_statistic = np.dot(delta_params.T, np.linalg.inv(delta_cov)).dot(delta_params)

    df = delta_params.size

    p_value = 1 - sts.chi2.cdf(hausman_statistic, df)

    print(f'Значение статистики: {hausman_statistic:.4f}')
    if p_value < alpha:
        print(f'P-значение: {p_value:.4f} < {alpha} => выбираем модель с фиксированными эффектами')
    else:
        print(f'P-значение: {p_value:.4f} > {alpha} => выбираем модель со случайными эффектами')

Hausman_test(fe_res, re_res, 0.05)
```

```python
test_regions = df_test.reset_index()['Название субъекта'].tolist()

# region = 'Респ. Дагестан'
for region in test_regions:
  df_train_new = df_train.reset_index()
  real_train = df_train_new[df_train_new['Название субъекта'] == region]['Ожидаемая продолжительность жизни  при рождении, число лет:']

  X_train_new = X_train.drop(columns=['Среднедушевые денежные доходы  (в месяц), руб.', 'Уровень безработицы, процентов',
                                                    'Численность среднего медицинского  персонала, человек (на 10 000 человек населения)'])
  X_test_new = X_test.drop(columns=['Среднедушевые денежные доходы  (в месяц), руб.', 'Уровень безработицы, процентов',
                                                    'Численность среднего медицинского  персонала, человек (на 10 000 человек населения)'])

  est_eff_reg = re_res.estimated_effects.loc[(region)].values[0][0]
  estimated_train = (re_res.predict(sm.add_constant(X_train_new)) + est_eff_reg).values

  forecast = (re_res.predict(sm.add_constant(X_test_new, has_constant='add')) + est_eff_reg).values


  years = np.arange(2010, 2022)

  plt.figure(figsize=(10, 6))
  plt.plot(years, real_train, label='Фактические значения', marker='o')
  plt.plot(years[:len(estimated_train)], estimated_train, label='Оцененные значения', linestyle='--', marker='x')
  plt.plot(2022, forecast, label='Прогнозное значения', linestyle='--', marker='o')


  # Set x-ticks for every year
  plt.xticks(np.arange(2010, 2025, 1))

  # Set x-axis limit to 2024
  plt.xlim(2009, 2024)
  plt.ylim(50, 100)


  plt.xlabel('Год')
  plt.ylabel('Коэффициенты естественного прироста')
  plt.title(f'Регион: {region}')
  plt.legend(loc='upper left')
  plt.grid(True)
  plt.show()
```

C помощью модели с фиксированными эффектами были рассчитаны прогнозные значения ОПЖ для 7 субъектов СКФО РФ.  
Сравнение реальных значений с прогнозными показало, что для Дагестана, Ингушетии, КЧР, КБР и Чечни предсказанные значения хорошо согласуются с фактическими данными. В Северной Осетии и Ставропольском крае согласие немного хуже, но направление тренда предсказано правильно. Это свидетельствует о высокой прогностической способности выбранной модели.
